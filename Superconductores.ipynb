```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Obtén los datos:
import pandas as pd

df = pd.read_csv('datos.txt', sep='\t')

#print(df)
```

```python
 
df.plot(x='Temperature (K)', y='Long Moment (emu)')


plt.legend()
plt.show()
```

```python
#momento/masa/campo de prueba
#MASA m=0.0985g
#CAMPO DE PRUEBA 100Oe

K=0.0985/100
df['Long Moment (emu)']/df['Field (Oe)']/0.0985
df.plot(x='Temperature (K)', y='Long Moment (emu)')

plt.ylabel('Susceptibilidad Magnética')
plt.show()
```

```python
percent=[]
temperature=[]

for i in range(0,26):
    print("25-i:", 25-i,"; 25+i:", 25+i)
    percent.append(df['Long Moment (emu)'].iloc[25-i]*100/df['Long Moment (emu)'].iloc[25+i])
    print("percent?",df['Long Moment (emu)'].iloc[25-i]*100/df['Long Moment (emu)'].iloc[25+i])
    temperature.append((df['Temperature (K)'].iloc[25-i]+df['Temperature (K)'].iloc[25+i])/2)
    print("temp:",(df['Temperature (K)'].iloc[25-i]+df['Temperature (K)'].iloc[25+i])/2)


plt.plot(temperature, percent, '-gD', markevery=[5,6,7])
plt.ylabel('%')
plt.xlabel('Temperatura (K)')
plt.title('Porcentaje de Fill-Pulling')


print("--PICO INTERESANTE--")
for i in range(0,3):
    print("TEMP",temperature[5+i])
    print("PERCENT",percent[5+i])
```
